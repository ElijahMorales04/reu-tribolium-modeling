## 1D BIFURCATION

```python
import sympy as sp
import numpy as np
import matplotlib.pyplot as plt

# Define symbolic variables
x, r = sp.symbols('x r')

# Define the equation dxdt = rx - ln(1 + x)
dxdt = r*x - sp.log(1 + x)

# Define the function dsdt
def dsdt(x_val, r_val):
    return r_val * x_val - np.log(1 + x_val)

# Generate an array of values for x
x_values = np.linspace(-3, 3, 100)

# Define different values for the parameter r
r_values = [1, 0, -1]

# Create the plot
plt.figure()

# Plot for each value of r
for r_val in r_values:
    plt.plot(x_values, dsdt(x_values, r_val), linewidth=1, label=r'$r = {}$'.format(r_val))

# Set the limits of the axes
plt.xlim([-3, 3])
plt.ylim([-10, 10])

# Add a legend and axis labels
plt.legend(loc='lower right')
plt.title(r'$\dot{x} = rx - \ln(1 + x)$', fontsize=14)
plt.xlabel(r'$x$', fontsize=14)
plt.ylabel(r'$\dot{x}$', fontsize=14)

# Add a grid and display the plot
plt.grid(True)
plt.show()
```

```python
import sympy as sp

# Define symbolic variables
x, y, alpha = sp.symbols('x y alpha')

# Define the system of differential equations
dxdt = y
dydt = x**2 - y - alpha

# Define the system
system = sp.Matrix([dxdt, dydt])

# Solve the system of equations for x and y
solutions = sp.solve(system, (x, y))

# Extract the solutions for x and y
eq1 = (solutions[0][0], solutions[0][1])
eq2 = (solutions[1][0], solutions[1][1])

# Compute the Jacobian matrix
J = system.jacobian([x, y])

# Substitute the values from eq1 into the Jacobian matrix
J1 = J.subs([(x, eq1[0]), (y, eq1[1])])

# Compute the eigenvalues of J1
Eig1 = J1.eigenvals()

# Substitute alpha values into the eigenvalues
Eig1A1 = [e.subs(alpha, 1) for e in Eig1]
Eig1A0 = [e.subs(alpha, 0) for e in Eig1]
Eig1AN1 = [e.subs(alpha, -1) for e in Eig1]

# Repeat the process for eq2
J2 = J.subs([(x, eq2[0]), (y, eq2[1])])
Eig2 = J2.eigenvals()
Eig2A1 = [e.subs(alpha, 1) for e in Eig2]
Eig2A0 = [e.subs(alpha, 0) for e in Eig2]
Eig2AN1 = [e.subs(alpha, -1) for e in Eig2]

# Display the results
eq1, eq2, J, J1, Eig1A1, Eig1A0, Eig1AN1, J2, Eig2A1, Eig2A0, Eig2AN1
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint

# Define the system of differential equations
def example2(Y, t, alpha):
    x, y = Y
    dxdt = y
    dydt = x**2 - y - alpha
    return [dxdt, dydt]

# Define initial conditions and time span
X0, Y0 = -1, -0.5
inits = [X0, Y0]
tspan = np.arange(0, 20, 0.1)

# Define alpha
alpha = 1

# Solve the ODEs
solution = odeint(example2, inits, tspan, args=(alpha,))
solutionX = solution[:, 0]
solutionY = solution[:, 1]

# Plot X and Y against time
plt.figure()
plt.plot(tspan, solutionX, linewidth=1, label='X Vs T')
plt.plot(tspan, solutionY, linewidth=1, label='Y Vs T')
plt.legend()
plt.title('Plotted against time')
plt.xlabel('Time')
plt.ylabel('Value')
plt.grid(True)
plt.show()

# Plot phase plane
plt.figure()
plt.plot(solutionX, solutionY, linewidth=1)
plt.title('Phase plane')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
plt.show()

# Plot vector field for different alpha values
alpha_values = [1, 0, -1]
for alpha in alpha_values:
    X, Y = np.meshgrid(np.arange(-3, 3.4, 0.4), np.arange(-3, 3.4, 0.4))
    U = Y
    V = -Y + X**2 - alpha
    plt.figure()
    plt.quiver(X, Y, U, V)
    plt.title(r'$\alpha = {}$'.format(alpha))
    plt.xlabel('x')
    plt.ylabel('y')
    plt.grid(True)
    plt.xlim(-3, 3)
    plt.ylim(-3, 3)
    plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the model
def dx_dt(x, r):
    return r*x + x**3 - x**5

# Define a function to find the fixed points
def find_fixed_points(r):
    return np.roots([1, 0, 0, r, -1, 0])

# Define a function to determine stability
def determine_stability(x, r):
    stability = []
    for root in x:
        derivative = r + 3*root**2 - 5*root**4
        if derivative < 0:
            stability.append('Stable')
        elif derivative > 0:
            stability.append('Unstable')
        else:
            stability.append('Saddle')
    return stability

# Define range of parameter values
r_values = np.linspace(-3, 3, 400)

# Initialize lists to store bifurcation points and stability
bifurcation_points = []
stability = []

# Loop through each value of r
for r in r_values:
    # Find fixed points for the current value of r
    fixed_points = find_fixed_points(r)
    # Determine stability of fixed points
    stability.extend(determine_stability(fixed_points, r))
    # Concatenate fixed points lists for each r value
    bifurcation_points.extend(fixed_points)

# Plot bifurcation diagram
plt.figure(figsize=(8, 6))
plt.scatter(np.repeat(r_values, len(fixed_points)), bifurcation_points, c='b', s=5)
plt.title('Bifurcation Diagram')
plt.xlabel('r')
plt.ylabel('Fixed Points')
plt.grid(True)
plt.show()
```

```python

import sympy as sp
import numpy as np
import matplotlib.pyplot as plt

# Define symbols
x, r = sp.symbols("x r")

# Define the differential equation
dxdt = r * x + x**3 - x**5

# Solve for fixed points (dxdt = 0)
solnx = sp.solve(dxdt, x)

# Define the functions for substituting r
eq_funcs = [sp.lambdify(r, sol, "numpy") for sol in solnx]

# Generate an array of values for r
s = np.linspace(-1, 1, 400)

# Initialize lists to store the x-values of the fixed points
x_fixed_points = [[] for _ in range(len(eq_funcs))]

# Calculate the fixed points for each r
for r_val in s:
    for i, eq_func in enumerate(eq_funcs):
        try:
            sol = eq_func(r_val)
            if np.isreal(sol):
                x_fixed_points[i].append(np.real(sol))
            else:
                x_fixed_points[i].append(np.nan)
        except ValueError:
            x_fixed_points[i].append(np.nan)

# Convert to numpy arrays for plotting
x_fixed_points = [np.array(points) for points in x_fixed_points]

# Create the plot
plt.figure(figsize=(10, 6))

# Plot each set of fixed points
for i, points in enumerate(x_fixed_points):
    plt.plot(s, points, linewidth=1, label=f'Fixed Point {i+1}')

plt.grid(True)
plt.title(r'Bifurcation Diagram of $\dot{x} = rx + x^3 - x^5$', fontsize=14)
plt.legend(loc='best', fontsize=12)
plt.xlabel(r'$r$', fontsize=14)
plt.ylabel(r'$\hat{x}$', fontsize=14, rotation=0)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

def cubic_map(x, r):
    return r * x - x**3

# Function to find fixed points
def find_fixed_points(r):
    roots = np.roots([1, 0, -1, r])
    real_roots = roots[np.isreal(roots)].real
    return real_roots

# Function to check stability of fixed points
def check_stability(x, r):
    stable = []
    for xi in x:
        if abs(3 * xi**2 - r) < 1:
            stable.append(xi)
    return stable

# Define parameters
r_values = np.linspace(-2, 2, 1000)
x_values = np.linspace(-2, 2, 1000)

# Plot bifurcation diagram
plt.figure(figsize=(10, 6))
for r in r_values:
    stable_points = check_stability(find_fixed_points(r), r)
    plt.scatter([r] * len(stable_points), stable_points, s=0.1, color='blue')

plt.xlabel('r')
plt.ylabel('Fixed Points')
plt.title('Bifurcation Diagram of Cubic Map')
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

def planar_map(x, y, mu):
    xn1 = x + mu * y - x**2
    yn1 = x
    return xn1, yn1

# Function to find fixed points
def find_fixed_points(mu):
    roots = np.roots([1, -1, mu])
    return roots

# Function to check stability of fixed points
def check_stability(roots, mu):
    stable = []
    for root in roots:
        if root > 0 and root < mu:
            stable.append(root)
    return stable

# Define parameters
mu_values = np.linspace(0, 2, 1000)
x_values = np.linspace(-2, 2, 1000)

# Plot bifurcation diagram
plt.figure(figsize=(10, 6))
for mu in mu_values:
    roots = find_fixed_points(mu)
    stable_points = check_stability(roots, mu)
    plt.scatter([mu] * len(stable_points), stable_points, s=0.1, color='blue')

plt.xlabel('μ')
plt.ylabel('Fixed Points')
plt.title('Bifurcation Diagram of Planar Map')
plt.grid(True)
plt.show()
```
